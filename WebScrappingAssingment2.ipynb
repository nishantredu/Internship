```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Question 1 Solution
url = 'https://www.naukri.com/'
def get_data_analyst_bangalore(url):
    driver = webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    search_job = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[3]/div/div/div[1]/div/div/div/input')
    search_job.send_keys('Data Analyst')
    search_location = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[3]/div/div/div[5]/div/div/div/input')
    search_location.send_keys('Bangalore')
    driver.implicitly_wait(5)
    search_btn = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[3]/div/div/div[6]')
    search_btn.click()
    job_titles = driver.find_elements_by_xpath("//*[@class='title fw500 ellipsis']")
    driver.implicitly_wait(10)
    titles = []
    for j,i in enumerate(job_titles):
        titles.append(i.text)
        if j==9:
            break
    company_names = []
    companies = driver.find_elements_by_xpath("//*[@class='subTitle ellipsis fleft']")
    for j,i in enumerate(companies):
        company_names.append(i.text)
        if j==9:
            break
    
    ratings = []
    rating_job = driver.find_elements_by_xpath("//*[@class='starRating fleft dot']")
    for j,r in enumerate(rating_job):
        ratings.append(r.text)
        if j==9:
            break
    
    salaries = []
    salary_job = driver.find_elements_by_xpath("//*[@class='fleft grey-text br2 placeHolderLi salary']")
    for j,s in enumerate(salary_job):
        salaries.append(s.text)
        if j==9:
            break
        
    roles = []
    roles_job = driver.find_elements_by_xpath("//*[@class='tags has-description']")
    for j,r in enumerate(roles_job):
        roles.append(r.text.replace('\n', ', '))
        if j==9:
            break
        
    df = pd.DataFrame({'Title':titles,
                      'Company':company_names,
                      'Rating':ratings,
                      'Salary':salaries,
                      'Roles':roles})
    return df
    
```

```python
# Question 8 Solution
def get_amazon_laptop(url):

    driver = webdriver.Chrome('chromedriver.exe')
    driver.implicitly_wait(50)
    driver.get(url)
    driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]').send_keys('laptop')
    driver.find_element_by_xpath('//*[@id="nav-search-submit-button"]').click()
    driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a').click()
    products = []
    prices =[]
    ratings = []
    c=1
    for i in range(len(driver.find_elements_by_xpath('//*[@class="a-size-medium a-color-base a-text-normal"]'))):
        products.append(driver.find_elements_by_xpath('//*[@class="a-size-medium a-color-base a-text-normal"]')[i].text)
        prices.append(driver.find_elements_by_xpath('//*[@class="a-price-whole"]')[i].text)
        ratings.append(driver.find_elements_by_xpath('//*[@class="a-icon-alt"]')[i].text)      
        c+=1
        if c==10:
            break

    df = pd.DataFrame({
        'Title':products,
        'Price':prices,
        'Rating':ratings
    })
    return df
```

```python
url = 'http://www.ambitionbox.com/'
# Question 9 Solution
def get_ambitionBox_dataScienceNoida(url):
    driver = webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    driver.find_element_by_xpath('//*[@id="headerWrapper"]/nav/nav/a[6]').click()
    driver.find_element_by_xpath('//*[@id="jobs-typeahead"]/span/input').send_keys('Data Scientist')
    driver.find_element_by_xpath('//*[@id="jobs"]/div[2]/div[1]/div[1]/div/div/div/button').click()
    driver.find_element_by_xpath('//*[@id="filters-row"]/div/div/div[2]/div[1]/i').click()
    driver.find_element_by_xpath('//*[@id="filters-row"]/div/div/div[2]/div[2]/div/div[2]/input').send_keys('Noida')
    driver.find_element_by_xpath('//*[@id="filters-row"]/div/div/div[2]/div[2]/div/div[3]/div[1]/div[1]/div/label').click()
    designations = []
    companies = []
    experience = []
    exp = []
    salary =[]
    skills = []

    for i in range(10):
        designations.append(driver.find_elements_by_xpath('//*[@class="title noclick"]')[i].text)
        companies.append(driver.find_elements_by_xpath('//*[@class="company body-medium"]')[i].text)
        experience.append(driver.find_elements_by_xpath('//*[@class="other-info"]')[i].text)

    for i in experience:
        exp.append(i.split('\n')[0])
        salary.append(i.split('\n')[1])
        skills.append(i.split('\n')[2])

    df = pd.DataFrame({
        'Designation':designations,
        'Company':companies,
        'Experiece':exp,
        'Salary':salary,
        'Skills':skills
    })
    return df
```

```python
url = 'http://www.ambitionbox.com/'
# Question 10 Solution
def get_data_scientist_slary(url):
    driver = webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    driver.find_element_by_xpath('//*[@id="headerWrapper"]/nav/nav/a[4]').click()
    driver.find_element_by_xpath('//*[@id="jobProfileSearchbox"]').send_keys('Data Scientist')
    driver.find_element_by_xpath('//*[@id="salaries"]/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p').click()
    company_names = []
    experiences=[]
    average_ctc = []
    lower_sal = []
    upper_sal=[]
    for i in range(10):
        curr_company = driver.find_elements_by_xpath('//*[@class="company-info"]')[i].text
        company_names.append(curr_company.split('\n')[0])
        experiences.append(curr_company.split('\n')[2])
        average_ctc.append(driver.find_elements_by_xpath('//*[@class="averageCtc"]')[i].text)
        salary = driver.find_elements_by_xpath('//*[@class="salary-values"]')[i].text
        lower_sal.append(salary.split('\n')[0])
        upper_sal.append(salary.split('\n')[1])
    df = pd.DataFrame({
        'Company':company_names,
        'Experience':experiences,
        'Average CTC':average_ctc,
        'Lowest CTC':lower_sal,
        'Highest CTC':upper_sal
    })
    return df
```
